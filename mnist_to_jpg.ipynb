```python
# coding=utf-8
import numpy as np
import cv2
import os
```

```python
def save_mnist_to_jpg(mnist_image_file, mnist_label_file, save_dir):
    if 'train' in os.path.basename(mnist_image_file):
        num_file = 60000
        prefix = 'train'
        print("file_num is ",num_file)
    else:
        num_file = 10000
        prefix = 'test'
        print("file num is ",num_file)
    with open(mnist_image_file, 'rb') as f1:
        image_file = f1.read()

    with open(mnist_label_file, 'rb') as f2:
        label_file = f2.read()
    #print(len(label_file))
    image_file = image_file[16:]
    print(len(image_file))

    label_file = label_file[8:]
    print(len(label_file))
    print("The first number is ",label_file[0])
    print("The second number is ",label_file[1])
    for i in range(num_file):
        label_name = label_file[i]
        image_list = [int(item) for item in image_file[i * 784: i*784+784]] # Convert binary data to decimal
        image_array = np.array(image_list,dtype=np.uint8).reshape(28,28,1)
        print(image_array.shape)
        image_name = os.path.join(save_dir,str(i+1)+'_'+str(label_name)+'.jpg')
        cv2.imwrite(image_name,image_array)
        pass
```

```python
if __name__ == '__main__':
    test_image_file = 'data/digits/t10k-images-idx3-ubyte'
    test_label_file = 'data/digits/t10k-labels-idx1-ubyte'
    
    train_image_file = 'data/digits/train-images-idx3-ubyte'
    train_label_file = 'data/digits/train-labels-idx1-ubyte'
    
    save_train_dir ='data/digits/train_images/'
    save_test_dir ='data/digits/test_images/'

    if not os.path.exists(save_test_dir):
        os.makedirs(save_test_dir)
    
    if not os.path.exists(save_train_dir):
        os.makedirs(save_train_dir)

    save_mnist_to_jpg(test_image_file, test_label_file, save_test_dir)
    save_mnist_to_jpg(train_image_file, train_label_file, save_train_dir)
    
    
```
